```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
gen=pd.read_csv('IndividualDetails.csv')
gen
```

```python
gen_affected=gen.groupby('detected_state')['gender'].count()
gen_affected
```

```python
gen_1=gen[gen['gender']=='M']
gen_2=gen_1.groupby('detected_state')['gender'].count()
gen_3=gen[gen['gender']=='F']
gen_4=gen_3.groupby('detected_state')['gender'].count()
gen_4
```

```python
gen_2_prob=(gen_2/gen_affected)*100
gen_2_prob.reset_index()
```

```python
gen_4_prob=(gen_4/gen_affected)*100
gen_4_prob.reset_index()
```

```python
male=gen_2_prob.to_frame()
male.rename(columns={'gender':'Male probability'},inplace=True)
male
```

```python
female=gen_4_prob.to_frame()
female.rename(columns={'gender':'Female probability'},inplace=True)
female
```

```python
gender=female.merge(male,on='detected_state').reset_index()
gender
```

```python
trace1=go.Bar(x=gender['detected_state'],y=gender['Male probability'], name='male',
              marker={'color':'#000080'})


trace2=go.Bar(x=gender['detected_state'],y=gender['Female probability'], name='female',
              marker={'color':'#00FFFF'})

data=[trace1,trace2]

layout=go.Layout(title='COVID-19 affected (GENDER-WISE)',
                xaxis={'title':'STATES'},
                yaxis={'title':'PROBABILITY'})

fig=go.Figure(data=data, layout=layout)

pyo.plot(fig,filename='gender_graph.html')
```

```python
age=pd.read_csv('AgeGroupDetails.csv')
age
```

```python
age_final=age.groupby('AgeGroup')['Percentage'].sum().reset_index()
age_final
```

```python
trace=go.Bar(x=age_final['AgeGroup'],y=age_final['Percentage'],
              marker={'color':'#00FFFF'})


data=[trace]

layout=go.Layout(title='COVID-19 affected (AGE-WISE)',
                xaxis={'title':'Age Groups'},
                yaxis={'title':'Percentage Affected'})

fig=go.Figure(data=data, layout=layout)

pyo.plot(fig,filename='Age_graph.html')
```
